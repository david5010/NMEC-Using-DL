```python
# data = pd.read_csv('1minute_data_newyork/1minute_data_newyork.csv')
# data
data = pl.read_csv('data/15minute_data_austin/15minute_data_austin.csv')
data
```

```python
data = data.with_columns(pl.col('local_15min').s)
```

```python
# df = data.with_columns(
#     pl.col("local_15min").str.strptime(pl.Date32, "%Y-%m-%d %H:%M").alias("local_15min")
# )
df = data
columns_to_convert = [name for name in df.columns if name not in ["dataid", "local_15min"]]

for col_name in columns_to_convert:
    df = df.with_columns(
        pl.when(pl.col(col_name).is_null() | (pl.col(col_name).str.length() == 0))
        .then(0)
        .otherwise(pl.col(col_name).cast(pl.Float64))
        .alias(col_name)
    )

# Display the DataFrame
print(df)
```
